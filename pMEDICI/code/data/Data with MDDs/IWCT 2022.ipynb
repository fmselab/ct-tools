```python
import matplotlib.pyplot as plt
import numpy as np

labels_x = ["UNIFORM_BOOLEAN_1", "ADD_BOOLC_4"]
labels_x2 = ["MCA_1", "ADD_MCAC_2"]
n_threads = ["2 Threads", "4 Threads", "6 Threads", "8 Threads"]
times_1A = [0.011, 0.011, 0.012, 0.012]
times_2A = [5.164, 5.896, 5.667, 5.396]
times_2B = [29.763, 28.645, 27.182, 27.030]
times_1B = [0.091, 0.102, 0.092, 0.097]

sizes_1A = [6, 7, 7, 9]
sizes_2A = [2735, 2502, 2268, 2235]
sizes_2B = [3233, 3048, 2877, 2617]
sizes_1B = [12, 14, 13, 12]
```

```python
data1 = [times_1A, times_1B]
x = np.arange(len(n_threads))
width = 0.20

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(x - width/2, data1[0], color = '#B4BBC4', width = width)
ax.bar(x + width/2, data1[1], color = '#000000', width = width)
ax.legend(labels=labels_x)
ax.set_xticks(x)
ax.set_xticklabels(n_threads)
ax.set_ylabel('[seconds]')
```
